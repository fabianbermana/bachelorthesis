```python
import pandas as pd
import numpy as np
import sklearn as sk
from sklearn import metrics

import time
```

```python
# loading Goyal data
GoyalData_NBER = pd.read_csv("GoyalData_2021.csv", thousands=',')

# starting and ending periods
start = 192612
end = 201612

# keep only necessary time periods
start_index = GoyalData_NBER[GoyalData_NBER['yyyymm'] == start].index[0]
end_index = GoyalData_NBER[GoyalData_NBER['yyyymm'] == end].index[0] + 1

data = GoyalData_NBER

TRAIN_START = 192612
TRAIN_END = 195612
TEST_START = 195701
TEST_END = 201612

IND_TRAIN_START = data[data['yyyymm'] == TRAIN_START].index[0]
IND_TRAIN_END = data[data['yyyymm'] == TRAIN_END].index[0] + 1
IND_TEST_START = data[data['yyyymm'] == TEST_START].index[0]
IND_TEST_END = data[data['yyyymm'] == TEST_END].index[0] + 1

data
```

```python
data_stocks = data.copy()[['yyyymm']]

data_stocks['EQPREM_+1'] = data.loc[:,'CRSP_SPvw'].shift(-1)

data_stocks['DP'] = np.log(data['D12'] / data['Index'])

data_stocks['DY'] = np.log(data['D12'] / data['Index'].shift(1))

data_stocks['DE'] = np.log(data['D12'] / data['E12'])

data_stocks['EP'] = np.log(data['E12'] / data['Index'])

data_stocks['SVAR'] = data['svar']

data_stocks['BM'] = data['b/m']

data_stocks['NTIS'] = data['ntis']

data_stocks['TBL'] = data['tbl']

data_stocks['LTY'] = data['lty']

data_stocks['LTR'] = data['ltr']

data_stocks['TMS'] = data['lty'] - data['tbl']

data_stocks['DFY'] = data['BAA'] - data['AAA']

data_stocks['DFR'] = data['corpr'] - data['ltr']

data_stocks['INFL'] = data['infl']

data_stocks = data_stocks.drop(columns=['LTY', 'DE'])

data_stocks.iloc[IND_TRAIN_START:IND_TEST_END,:].to_csv('data_e.csv', index=False)

data_stocks
```

```python
from sklearn.preprocessing import StandardScaler

data_train = data_stocks.iloc[IND_TRAIN_START:IND_TRAIN_END-1,:]
data_test = data_stocks.iloc[IND_TEST_START-1:IND_TEST_END-1,:]

X_train = np.array(data_train.drop(columns=['yyyymm', 'EQPREM_+1']))
X_test = np.array(data_test.drop(columns=['yyyymm', 'EQPREM_+1']))
y_train = np.reshape(np.array(data_train.loc[:,['EQPREM_+1']]), (X_train.shape[0],))
y_test = np.reshape(np.array(data_test.loc[:,['EQPREM_+1']]), (X_test.shape[0],))

X_scaled_train = X_train.copy()
X_scaled_train[:,:] = StandardScaler().fit(X_train).transform(X_scaled_train)

X_scaled_test = X_test.copy()
X_scaled_test[:,:] = StandardScaler().fit(X_test).transform(X_scaled_test)
```

```python
def R2OS(MSFE_M, MSFE_bmk):
    return 1 - (MSFE_M / MSFE_bmk)
```

```python
# historical average - base model
historicalaverage = np.mean(y_train)
pred_historicalaverage = np.array([historicalaverage] * y_test.shape[0])

MSPE_bmk = sk.metrics.mean_squared_error(y_test, pred_historicalaverage)
R2OS_bmk = R2OS(MSPE_bmk, MSPE_bmk)

print(MSPE_bmk, R2OS_bmk)
```

```python
# kitchen sink
from sklearn.linear_model import LinearRegression

kitchensink = LinearRegression().fit(np.array(X_train), np.array(y_train))
pred_kitchensink = kitchensink.predict(X_test)

MSPE_kitchensink = metrics.mean_squared_error(y_test, pred_kitchensink)
R2OS_kitchensink = R2OS(MSPE_kitchensink, MSPE_bmk)

print(MSPE_kitchensink, R2OS_kitchensink)
```

```python
# LASSO
from sklearn.linear_model import LassoCV

lasso = LassoCV().fit(np.array(X_train), np.array(y_train))
pred_lasso = lasso.predict(X_test)

MSPE_lasso = metrics.mean_squared_error(y_test, pred_lasso)
R2OS_lasso = R2OS(MSPE_lasso, MSPE_bmk)

print(MSPE_lasso, R2OS_lasso)
```

```python
# elastic net
from sklearn.linear_model import ElasticNetCV
    
elasticnet = ElasticNetCV().fit(np.array(X_train), np.array(y_train))
pred_elasticnet = elasticnet.predict(X_test)

MSPE_elasticnet = metrics.mean_squared_error(y_test, pred_elasticnet)
R2OS_elasticnet = R2OS(MSPE_elasticnet, MSPE_bmk)

print(MSPE_elasticnet, R2OS_elasticnet)
```

```python
from sklearn.linear_model import LassoCV, ElasticNetCV
from sklearn.model_selection import TimeSeriesSplit
from sklearn import metrics

def expanding_window(X_train, X_test, y_train, y_test, forecast_model, model_args=None):
    """
    parameters
    -----------
    X_train: pandas DataFrame, independent variables used to train the forecast model
    X_test: pandas DataFrame, independent variables used for forecasting
    y_train: pandas DataFrame, dependent variable used to train the forecast model
    y_test: pandas DataFrame, dependent variable to forecast
    forecast_model: python object, model used for forecasting. Make this object similar to the
    sklearn models, with a fit() and predict() function
    model_args: dict, arguments to pass when initializing the model
    """
    start_time = time.time()
    if model_args is None:
        model_args = {}
    n_periods = y_test.shape[0]
    pred_model = np.zeros((n_periods,))
    pred_bmk = np.zeros((n_periods,))
    
    # run a forecast for every test period
    for i in range(n_periods):
        # forecasts of model
        model = forecast_model(**model_args).fit(X_train, y_train)
        pred = model.predict(X_test[[i],:])
        pred_model[i] = pred
        
        # forecasts of historical average
        bmk = np.mean(y_train)
        pred_bmk[i] = bmk
        
        # add new row to training data
        X_train = np.vstack([X_train, X_test[[i],:]])
        y_train = np.hstack([y_train, y_test[i]])
    
    MSPE_model = metrics.mean_squared_error(y_test, pred_model)
    MSPE_bmk = metrics.mean_squared_error(y_test, pred_bmk)
    R2OS_val = R2OS(MSPE_model, MSPE_bmk)
    
    end_time = time.time()
    elapsed_time = end_time - start_time

    return {'out-of-sample R2': R2OS_val, 'MSPE_bmk': MSPE_bmk, 'elapsed time': elapsed_time}


cv = TimeSeriesSplit(n_splits=5)
#print(expanding_window(X_train, X_test, y_train, y_test, LinearRegression, model_args={'fit_intercept': False}))
print(expanding_window(X_scaled_train, X_scaled_test, y_train, y_test, LassoCV, model_args={'cv': cv}))
#print(expanding_window(X_train, X_test, y_train, y_test, ElasticNetCV, model_args={'cv': cv}))
```
